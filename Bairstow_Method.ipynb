```python
def calc_b(r,s,a):
    b=[]
    for i, e in reversed(list(enumerate(a))):
        print("i="+str(i))
    
        if(i==int(len(a))-1):
            b=[a[i]]+b
        elif i==int(len(a))-2:
            b=[a[i]+r[-1]*b[0]]+b
        else:
            b=[a[i]+r[-1]*b[0]+s[-1]*b[1]]+b
        print("bnext"+str(b[0]))
    return b
    
```

```python
def calc_c(r,s,b):
    c=[]
    for i, e in reversed(list(enumerate(a))):
        print("i="+str(i))

        if(i==int(len(a))-1):
            c=[b[i]]
        elif i==int(len(a))-2:
            c=[b[i]+r[-1]*c[0]]+c
        else:
            c=[b[i]+r[-1]*c[0]+s[-1]*c[1]]+c
        print('cnext='+str(c[0]))
    return c
```

```python
itera=20

for i in range(0,itera):
    b=calc_b(r,s,a)
    c=calc_c(r,s,b)
    del_s=(b[1]*c[1]-b[0]*c[2])/(c[2]*c[2]-c[1]*c[3])
    print("del_s="+str(del_s))
    del_r=(b[0]*c[1]*c[3]-c[2]*b[1]*c[1])/(c[2]*c[2]-c[1]*c[3])
    print("del_r="+str(del_r))
    r.append(r[-1]+del_r)
    s.append(s[-1]+del_s)
    print("rnew="+str(r[-1]))
    print("snew="+str(s[-1]))
print(r[-1],s[-1])
```
