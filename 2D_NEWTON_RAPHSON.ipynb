```python
#------------Python plot Template (delet if not required)
import prutorlib as pl
import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 5, 3
# figure of the size 5in x 3in
def f1(x,y):
    return -3*x+ y +1
def f2(x,y):
    return x*x+36*y*y-36
def delf1_delx(x,y):
    epsilon=10^(-8)
    return (f1(x+epsilon,y)-f1(x,y))/epsilon
def delf1_dely(x,y):
    epsilon=10^(-8)
    return (f1(x,y+epsilon)-f1(x,y))/epsilon
def delf2_delx(x,y):
    epsilon=10^(-8)
    return (f2(x+epsilon,y)-f2(x,y))/epsilon
def delf2_dely(x,y):
    epsilon=10^(-8)
    return (f2(x,y+epsilon)-f2(x,y))/epsilon


x0=float(input('Initial x value'))
y0=float(input('Initial y value'))
threshold=0.001   #percentage
iteration=100   #iterations
iter_array=[]   #iteration storage
error=[]   #error storage
f=0
for i in range(0,iteration):
    d1_x=delf1_delx(x0,y0)
    d2_x=delf2_delx(x0,y0)
    d1_y=delf1_dely(x0,y0)
    d2_y=delf2_dely(x0,y0)
    Det=d1_x*d2_y-d2_x*d1_y
    x1=x0-((f1(x0,y0)*d2_y-f2(x0,y0)*d1_y))/Det
    y1=y0-(f2(x0,y0)*d1_x-f1(x0,y0)*d2_x)/Det
    if i>0:
        err=abs(((f1(x0,y0)*d2_y-f2(x0,y0)*d1_y)/Det)/x1)
        err2=abs(((f2(x0,y0)*d1_x-f1(x0,y0)*d2_x)/Det)/y1)
        error=error+[max(err,err2)]
        iter_array=iter_array+[i+1]
    x0=x1
    y0=y1
    if i>0 and error[-1]*100<threshold:
        f=1
        print('')
        print('Error became less than 0.001%')
        print('x=',x0)
        print('y=', y0)
        break
    
if f==0:
    print('')
    print('Iterations completed')
    print('x=',x0)
    print('y=', y0)

print('error=',error)
print('iter=',iter_array)
x=iter_array
y=error
plt.plot(x,y, 'r.', label = r'$y=f(x)$') 
plt.axhline(0, color='k')
plt.axvline(0, color='k')
# plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$y$', fontsize=20)
plt.legend(loc=4)
plt.tight_layout()

pl.prutorsaveplot(plt, 'plot2d.pdf')
```
